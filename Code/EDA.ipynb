# Import necessary libraries and load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

# Get an overview of the dataset with basic statistics and info

```python
# Display general information about the dataset
data.info()

# Show summary statistics for numerical columns
data.describe()
```

# Check for missing values in the dataset

```python
# Display the count of missing values in each column
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

# Visualize the distribution of the target variable 'label'

```python
plt.figure(figsize=(6,4))
sns.countplot(x='label', data=data)
plt.title('Distribution of Labels')
plt.xlabel('Label (0 = Human, 1 = Machine)')
plt.ylabel('Count')
plt.show()
```

# Visualize the distributions of some readability metrics

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['automated_readability_index'], kde=True, bins=30)
plt.title('Distribution of Automated Readability Index')
plt.xlabel('Automated Readability Index')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['flesch_reading_ease'], kde=True, bins=30)
plt.title('Distribution of Flesch Reading Ease')
plt.xlabel('Flesch Reading Ease')
plt.ylabel('Frequency')
plt.show()
```

# Correlation heatmap to check relationships between numerical features

```python
# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(14, 10))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
